```python
import pickle
import snap
```

```python
def get_comm_info(comm_file):
    '''
    get community information, two maps
    map1: key: user id, value: community id array
    map2: key: community id, value: user id array
    '''
    comm_map_usr = {}
    comm_map_comm = {}
    comm_id = 0
    with open(comm_file, 'r') as cf:
        for line in cf:
            node_list = line.split('\t')
            node_list = [int(id) for id in node_list]
            for id in node_list:
                if id in comm_map_usr:
                    comm_map_usr[id].append(comm_id)
                else:
                    comm_map_usr[id] = [comm_id]
            comm_map_comm[comm_id] = node_list[:]
            comm_id += 1
    return comm_map_usr, comm_map_comm
```

```python
gf_file = "data/com-lj.ungraph.txt"
gf = snap.LoadEdgeList(snap.PUNGraph, gf_file, 0, 1)
```

```python
print "Load graph! With nodes ", gf.GetNodes(), " and edges ", gf.GetEdges()
```

```python
##--Degree centrality
degr_centr_map = {}

for ni in gf.Nodes():
    degr_centr_map[ni.GetId()] = snap.GetDegreeCentr(gf, ni.GetId())
```

```python
degr_centr_file = "deg_ctr.pkl"

with  open(degr_centr_file, "wb") as fl:
    pickle.dump(degr_centr_map, fl)
```

```python
##--Betweenness Centrality
ni_btw_c_ary = snap.TIntFltH()
edg_btw_c_ary = snap.TIntPrFltH()

snap.GetBetweennessCentr(gf, ni_btw_c_ary, edg_btw_c_ary)
```

```python
##--Get closeness centrality
cl_centr_map = {}

for ni in gf.Nodes():
    cl_centr_map[ni.GetId()] = snap.GetClosenessCentr(gf, ni.GetId())
```

```python
cl_centr_file = "cl_ctr.pkl"

with  open(cl_centr_file, "wb") as fl:
    pickle.dump(cl_centr_map, fl)
```

```python
##--Get farness centrality
fr_centr_map = {}

for ni in gf.Nodes():
    fr_centr_map[ni.GetId()] = snap.GetFarnessCentr(gf, ni.GetId())
```

```python
##--Get page rank centrality (may need to calibrate the parameter)
pgrk_H = snap.TIntFltH()

snap.GetPageRank(gf, pgrk_H)
```
